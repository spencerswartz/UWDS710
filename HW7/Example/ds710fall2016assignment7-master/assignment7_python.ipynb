#Problem 1(a).  Word length.

Write a function called ```word_length_list()``` which takes a string and returns a list with the length of each word in the string.  Your function should ignore punctuation.  For example, ```word_length_list("The panda eats, shoots, and leaves!")``` should return ```[3,5,4,6,3,6]```.  

Call or create other functions as necessary to organize your work.

````python
def word_length_list(words_input):
  """ Takes a string and returns a list with the length of each word in the string """
  words_list = words_input.split()
  word_counts = []
  for word in words_list:
    word_counts.append(len(word))
  return word_counts
````

#Problem 1(c). _Pride and Prejudice_.

Use Python to count the number of words and mean length of words in each sentence of Pride and Prejudice (https://www.gutenberg.org/cache/epub/1342/pg1342.txt).  Save your result as a ```.csv``` file.

* Create new functions as necessary to organize your work.
* Include comments to explain the purpose and arguments of each function you create.

````python
"""Use Python to count the number of words and mean length of words in each 
sentence of Pride and Prejudice. Save your result as a .csv file."""
import re
import csv

# list of words whose abbreviations we do not want to count as a sentence
titles = "(Mr|Mrs|Ms|Dr)[.]"

sentence_lengths = "sentence_lengths.csv"
pride_and_prejudice_path = "Pride_And_Prejudice.txt"

def word_count_by_sentence(file_path):
  """ Find the number of words and the average word length of each sentence in a supplied txt file """
  # open the file
  with open(file_path) as f:
    file_text = f.read()
    # replace all new lines with spaces
    file_text = file_text.replace("\n", " ")
    # for each of the titles in titles, strip . instances
    # when it follows any of the strings preceding it in the declaration
    file_text = re.sub(titles, "\\1", file_text)
    # assuming that all remaining periods end a sentence
    # replace the period character with the string <split>
    file_text = file_text.replace(".", "<split>")
    # replace the ! character with <split>
    file_text = file_text.replace("!", "<split>")
    # replace the ? character with <split>
    file_text = file_text.replace("?", "<split>")
    # regex to remove all non-alphabetic and non-numeric characters as well as leaving <> and 'spaces' alone
    file_text = re.sub('[^A-Za-z0-9<> ]', '', file_text)
    # create a list delimited by <split>
    sentences = file_text.split("<split>")
    # strip the leading spaces off of all elements of the list
    sentences = [s.strip() for s in sentences]
    # truncate the last list entry since it is empty
    sentences = sentences[:-1]

    # Open csv file to write to
    with open(sentence_lengths, 'w') as csvfile:
      # Instantiate a writer and tell it the file to write to
      writer = csv.writer(csvfile, lineterminator = '\n')
      # Loop through the dictionary and write each key-value pair on its own row
      writer.writerow(["word_count", "avg_word_length"])
      for s in sentences:
        # make a list out of the words in the sentence
        word_list = s.split()
        # get the number of words in the sentence
        word_count = len(word_list)
        total_word_length = 0
        # loop through each word in the sentence
        for w in word_list:
          # get the total length of the sentence in terms of number of letters
          total_word_length = total_word_length + len(w)
        # find average length of each word in the sentence
        avg_word_length = total_word_length / float(word_count)
        # write the word cound and average word length of the sentence to the csv
        writer.writerow([word_count, avg_word_length])

word_count_by_sentence(pride_and_prejudice_path)
````
