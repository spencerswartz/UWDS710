```python
def gettweets(searchQuery,maxTweets,tweetsPerQry,sinceID,fName,max_id = 0):
    import sys
    import jsonpickle
    import os

    #searchQuery =  this is the string to search on.
    #maxTweets = Max tweets to return
    #tweetsPerQry = # of tweets per query
    #sinceId = (ex. 805464349745811456) twitter id this will be a lower bound in 
              #the querry so that not tweets befor this tweet are returned.
    #toID = end id not tweets after this id 
    #fName = the tweets will be stored here (ex. tweets.txt)
    #max_id = this is the to id as in no tweets from after this tweet default 0 if no limit needed

    # If results only below a specific ID are, set max_id to that ID.
    # else default to no upper limit, start from the most recent tweet matching the search query.
    

    tweetCount = 0
    print("Downloading max {0} tweets".format(maxTweets))
    with open(fName, 'w') as f:
        while tweetCount < maxTweets:
            try:
                if (max_id <= 0):
                    if (not sinceId):
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry)
                    else:
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                                since_id=sinceId)
                else:
                    if (not sinceId):
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                                max_id=str(max_id - 1))
                    else:
                        new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                                max_id=str(max_id - 1),
                                                since_id=sinceId)
                if not new_tweets:
                    print("No more tweets found")
                    break
                for tweet in new_tweets:
                    f.write(jsonpickle.encode(tweet._json, unpicklable=False) +
                            '\n')
                tweetCount += len(new_tweets)
                if tweetCount%500==0:
                    print("Downloaded {0} tweets".format(tweetCount))
                else:
                    None
                max_id = new_tweets[-1].id
            except tweepy.TweepError as e:
                # Just exit if any error
                print("some error : " + str(e))
                break

    print("Downloaded {0} tweets, Saved to {1}".format(tweetCount, fName))
    return None
```
